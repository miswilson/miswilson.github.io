```python
def make_term_search(driver, term, start_date, end_date):

    search_box = driver.find_element(By.ID, "search-box")  
    search_box.clear()
    search_box.send_keys(term)

    start_date_box = driver.find_element(By.ID, "start-date")  
    start_date_box.clear()
    start_date_box.send_keys(start_date)
    
    end_date_box = driver.find_element(By.ID, "end-date")  
    end_date_box.clear()
    end_date_box.send_keys(end_date)

   
    search_button = driver.find_element(By.ID, "search-button")  
    search_button.click()
```

```python
start_date = pd.Timestamp('1900-01-01')
end_date = pd.Timestamp('2023-01-31')

date_range = pd.date_range(start=start_date, end=end_date)

mention_counts = {}

for date in date_range:

    date_str = date.strftime('%Y-%m-%d')
    
    make_term_search(driver, term, date_str, date_str)  # Assuming make_term_search accepts a single date range


    count_element = driver.find_element(By.CLASS_NAME, "count")
    count_text = count_element.text
    count = int(count_text.split("(")[1].split(")")[0])

    mention_counts[date_str] = count

print(mention_counts)
```
