```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep

# Define terms and date range
terms = ["suit"]
start_date = pd.to_datetime("1900-01-01")
end_date = pd.to_datetime("2024-01-01")
date_range = pd.date_range(start=start_date, end=end_date, freq='D')  # Daily frequency

# Initialize list to store the counts by date and term
mention_counts = []

# Open the WebDriver
driver = webdriver.Chrome()  # Ensure ChromeDriver is installed and in PATH

# Loop through each date in the range
for date in date_range:
    date_str = date.strftime('%Y-%m-%d')  # Format date as needed
    
    for term in terms:
        # Construct the URL with the search term
        url = f"https://archive.vogue.com/search?QueryTerm={term}"
        driver.get(url)

        # Set start and end date for the search filter on the page
        try:
            start_date_box = driver.find_element(By.ID, "start-date")  # Adjust if needed
            start_date_box.clear()
            start_date_box.send_keys(date_str)

            end_date_box = driver.find_element(By.ID, "end-date")  # Adjust if needed
            end_date_box.clear()
            end_date_box.send_keys(date_str)
            
            # Submit the date filter
            search_button = driver.find_element(By.CLASS_NAME, "submit-button")  # Adjust if needed
            search_button.click()

            # Allow time for results to load
            sleep(2)

            # Retrieve the count of mentions
            count_element = driver.find_element(By.CLASS_NAME, "count")  # Adjust locator as needed
            count_text = count_element.text
            count = int(count_text.split("(")[1].split(")")[0])  # Adjust parsing if format differs
        except:
            count = 0  # Set to zero if count is not found or date filter fails

        # Store the result
        mention_counts.append({"Date": date_str, "Term": term, "Mentions": count})

# Close the WebDriver
driver.quit()

# Convert results to DataFrame for easier viewing and manipulation
mention_counts_df = pd.DataFrame(mention_counts)
print(mention_counts_df)
```
